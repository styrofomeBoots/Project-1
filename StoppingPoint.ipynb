```python
dummies = pd.get_dummies(death05_15["marital_status"])
```

```python
complete = death05_15.join(dummies)
```

```python
YearGroup = complete.groupby("current_data_year")
```

```python
div = YearGroup["D"].sum()
wid = YearGroup["W"].sum()
mar = YearGroup["M"].sum()
sin = YearGroup["S"].sum()
unm = YearGroup["U"].sum()
```

```python
DeathByMarStat = pd.DataFrame({"Divorced": div,
                             "Widowed": wid,
                             "Married": mar,
                             "Single": sin,
                             "Unmarried": unm})
DeathByMarStat
```
